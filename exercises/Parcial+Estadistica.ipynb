## 1. ANOVA 1 v칤a
Un estudio reciente compar칩 los salarios iniciales de los nuevos graduados en diversos campos. Los resultados se presentan a continuaci칩n. A un nivel del 5%. 쯇arece que hay diferencia en los salarios promedio de los graduados en los diferentes campos?

|Graduado|||Campos de estudio||
|---|
||Finanzas|Mercadeo|Sistemas|Econom칤a|
|1|23.2|22.1|23.3|22.2|
|2|24.7|19.2|22.1|22.1|
|3|24.2|21.3|23.4|23.2|
|4|22.9|19.8|24.2|21.7|
|5|25.2|17.2|23.1|20.2|
|6|23.7|18.3|22.7|22.7|
|7|24.2|17.2|22.8|21.8|



```python
finanzas = [23.2, 24.7, 24.2, 22.9, 25.2, 23.7, 24.2] # Resultados tratamiento 1
mercadeo = [22.1, 19.2, 21.3, 19.8, 17.2, 18.3, 17.2] # Resultados tratamiento 2
sistemas = [23.3, 22.1, 23.4, 24.2, 23.1, 22.7, 22.8] # Resultados tratamiento 3
economia = [22.2, 22.1, 23.2, 21.7, 20.2, 22.7, 21.8] # Resultados tratamiento 4

n_campos = len(finanzas + mercadeo + sistemas + economia) # N칰mero total de observaciones
c_campos = 4 # N칰mero total de tratamientos

alpha_campos = 0.05 # Nivel de significancia
```

- Se plantea la hip칩tesis nula y alternativa:

**Ho:** 洧랞finanzas = 洧랞mercadeo = 洧랞sistemas = 洧랞economia

**Ha:** 洧녵洧녶 洧노洧녶洧녬洧녩洧 洧녳洧녩洧 洧녴洧뉧롐놿롐뒳롐뀛롐 洧멇롐럻롐 洧녰洧녮洧녹洧녩洧녳洧뉧롐

- Ahora encontraremos el valor cr칤tico para rechazar la hip칩tesis nula

```python
df1_campos = c_campos - 1
df2_campos = n_campos - c_campos
ms.hypothesis.crit_val_f(df1_campos, df2_campos, alpha_campos)
```

- Ahora encontraremos el estad칤stico de prueba (RAZ칍N F) para comparar con el valor cr칤tico y determinar si se rechaza la hip칩tesis nula

```python
ms.anova.get_fratio(finanzas, mercadeo, sistemas, economia)
```

Entonces, teniendo en cuenta que el valor de la RAZ칍N F **20.688556457065562**, es **mayor** al valor cr칤tico **3.0087865704473615**, se puede **rechazar con un nivel de significancia del *5%* la hip칩tesis nula.**

El estudio debe rechazar la hip칩tesis de que los salarios promedio de los graduados en los diferentes campos son iguales. Es decir, que estos salarios promedio **NO SON IGUALES** y **SI SE PRESENTAN DIFERENCIAS** entre ellos.

## 2. Pruebas de hip칩tesis de 1 cola para la media poblacional en muestras peque침as
Un nuevo bombillo producido por Sun Systems est치 dise침ado para incrementar la vida 칰til de los bombillos a m치s de 5000 horas que es el promedio de los que actualmente existen. 쮼l nuevo producto de Sun Systems proporciona una mejora si 25 bombillos se funden en promedio a las 5117 horas con s = 1886 horas? Fije alfa en 5%.

```python
mu_sun = 5000 # Hip칩tesis nula de la media poblacional
n_sun = 25 # Tama침o de la muestra
x__sun = 5117 # Promedio muestral
s_sun = 1886 # Desviaci칩n est치ndar muestral

alpha_sun = 0.05 # Nivel de significancia
```

<u> **Paso 1**</u>: Plantear hip칩tesis

**Ho:** 풮 = 5000                  

**Ha:** 풮 > 5000

<u> **Paso 2**</u>: Nivel de significancia

```python
alpha_sun
```

<u> **Paso 3**</u>: Valores cr칤ticos

```python
df_sun = n_sun - 1
crit_sun = ms.hypothesis.crit_val_t(df_sun, alpha_sun, 'right')
crit_sun
```

<u> **Paso 4**</u>: Estad칤stico de prueba (T)

```python
t_sun = ms.generals.get_t(x__sun, mu_sun, s_sun, n_sun)
t_sun
```

<u> **Paso 5**</u>: Decisi칩n

```python
ms.hypothesis.reject_h0(crit_sun, t_sun, 'right')
```

**NO se rechaza la hip칩tesis nula**, teniendo en cuenta que el estad칤stico de prueba *0.31018027571580065* no es mayor que el valor cr칤tico *1.7108820799094275*.

<u>**Paso 6**</u>: Conclusi칩n

Teniendo en cuenta un nivel de significancia del *5%*, podemos rechazar la hip칩tesis de que la duraci칩n promedio del nuevo bombillo producido por Sun Systems **es mayor** a 5000 horas.

## 4. Intervalos de confianza para la diferencia entre 2 medias poblacionales en muestras grandes
Dos procesos de producci칩n se utilizan para producir tubos de acero. Una muestra de 100 tubos tomada del primer proceso de producci칩n tiene una longitud en promedio de 27.3 pulgadas y s = 10.3 pulgadas. Las cifras correspondientes para los 100 tubos producidos por el segundo m칠todo son 30.1 y 5.2 . 쯈u칠 revela un intervalo del 99% sobre la diferencia de longitudes promedio de los tubos producidos por estos dos m칠todos?.

```python
n_p1 = 100 # Tama침o de la muestra 1
x__p1 = 27.3 # Promedio de la muestra 1
s_p1 = 10.3 # Desviaci칩n est치ndar de la muestra 1

n_p2 = 100 # Tama침o de la muestra 2
x__p2 = 30.1 # Promedio de la muestra 2
s_p2 = 5.2 # Desviaci칩n est치ndar de la muestra 2

alpha_p = 0.99 # Nivel de confianza
```

```python
ms.intervals.norm_2p(n_p1, n_p2, x__p1, x__p2, s_p1, s_p2, alpha_p)
```

**R/** A un nivel de confianza del *99%*, podemos observar que la diferencia entre la longitud promedio de los tubos producidos por estos 2 m칠todos est치 entre **-5.7720427057731767, 0.17204270577317526** pulgadas.

Entonces, como entre el intervalo de la diferencia est치 incluido el valor de **cero**, podemos concluir que la efectividad de estos 2 procesos de producci칩n pueden ser iguales, ya que no se encuentran suficientes evidencias para demostrar que un proceso sea mejor que el otro.
